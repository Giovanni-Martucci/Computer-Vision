# 1. Haralick Texture

<font size="3">Iniziamo con la classificazione delle texture tramite le features di Haralick. In particolare il concetto fondamentale coinvolto nel calcolo delle funzionalità di Haralick Texture è la matrice di ricorrenza del livello di grigio o GLCM.<br>
La matrice di ricorrenza del livello di grigio (GLCM) utilizza il concetto di adiacenza nelle immagini. L'idea di base è che cerca coppie di valori di pixel adiacenti che si verificano in un'immagine.
<br><br>Vi sono quattro tipi di adiacenza:<br><br>
- Da sinistra a destra;
- Dall'alto al basso;
- In alto da sinistra in basso a destra;
- Da destra in alto a sinistra in basso. </font>

<img src="./Matrix.png"/>

<font size="3"><br>Ecco un esempio del codice:<br>- Importiamo i packages necessari</font>

```python
import cv2
import numpy as np
import os
import glob
import mahotas as mt
from sklearn.svm import LinearSVC
import cvutils
```

<font size="3"> Carichiamo il training set ed inizializziamo due liste che conterranno i vettori di features e le etichette</font>

```python
train_path = "textH/dataset/train"
train_names = os.listdir(train_path)

train_features = []
train_labels = []
```

<br><font size="3">Implementiamo la funzione che andrà ad estrarre le features:</font>

```python
def extract_features(image):
    textures = mt.features.haralick(image)
    ht_mean  = textures.mean(axis=0)
    return ht_mean
```

<br><font size="3">Con la funzione *mt.features.haralick()* andiamo ad estrarre le features per tutti e 4 i tipi di adiacenza;<br><br>Andiamo poi a calcolare la media di tutti e 4 i tipi di GLCM, tramite *mean()*.<br><br> Infine, ritorniamo il vettore delle feautres il quale descrive la texture.<font>

<br><font size="3">Ora passiamo alla fase di training:<br><br>
<font size="3">
Scorrendo sulle etichette di allenamento, che abbiamo appena incluso nella directory di allenamento, prendiamo tutti i file con estensione jpg (per ogni etichetta alla volta) e analiziamo ogni file uno alla volta.<br><br>
Tramite le funzioni *cv2.imread()* e *cv2.cvtColor()* andiamo a leggere e convertire ciascuna immagine in scala di grigi.<br><br> Richiamiamo la funzione implementata precedentemente per estrarre le texture Haralick dall'immagine, per ogni immagine, in scala di grigi aggiungendoli alla lista *train_features* e aggiungendo l'etichetta in *train_labels*.
<font><font>

```python
print("[STATUS] Started extracting haralick textures..")
for train_name in train_names:
    cur_path = train_path + "/" + train_name
    cur_label = train_name
    i = 1

    for file in glob.glob(cur_path + "/*.jpg"):
        print("Processing Image - {} in {}".format(i, cur_label))
        image = cv2.imread(file)
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        # extract haralick texture from the image
        features = extract_features(gray)
        # append the feature vector and label
        train_features.append(features)
        train_labels.append(cur_label)
        # show loop update
        i += 1
```

<font size="3">Vediamo dal risultato che tutte le immagine sono state processate. (Nell'esempio abbiamo 5 immagini per ogni tipo di classe).</font>

<br><br><font size="3">Procediamo con la creazione del classificatore tramite la funzione *LinearSVC()* importata dalla libreria *sklearn*.<br><br>Di seguito i parametri che passiamo alla funzione:<br>- *random_state* impostato a 9: controlla la generazione di numeri pseudo casuali per mescolare i dati per la discesa a doppia coordinata;<br>- *tol*: indica la tolleranza per i criteri di arresto;<br>- *verbose* settato a 1: abilita l'output dettagliato;<br>- *max_iter*: indica il massimo numero di iterazioni da eseguire.</font>

```python
# create the classifier
print("[STATUS] Creating the classifier..")
clf_svm = LinearSVC(random_state=9,tol=1e-5, verbose=1, max_iter=100000)
```

<br><font size="3">Adesso si procede adattando le features e le labels di allenamento al classificatore, scegliendo un numero di iterazioni sufficienti affinchè il modello arrivi a convergenza</font>

```python
# fit the training data and labels
print("[STATUS] Fitting data/label to model..")
clf_svm.fit(train_features, train_labels)
```

<br><br><font size="3">Infine testiamo il classificatore, appena creato, sui dati di test:<br><br>
Inseriamo il path delle immagini di test in *test_path* e poi tramite il ciclo *for* effettueremo l'operazione su ogni immagine del dataset. Tramite la funzione di OpenCV *cv2.imread()* andreamo a caricare l'immagine corrente fornendo il path di essa,e si procede con la conversione in scala di grigi, tramite *cv2.cvtColor()* che accetta 2 parametri obbligatori:<br>
- src: l'immagine da trasformare.
- code: il codice relativo allo spazio colore da ottenere.<br>

Successivamente si estraggono le features dalle immagini in scala di grigi tramite la funzione precedentemente creata (*extract_features*). Dopodichè, tramite la funzione *predict()*, andremo a predirre l'etichetta di appartenenza.<br><br>Infine tramite *cv2.putText()* che accetta come parametri:<br>
- image:l'immagine su cui scrivere.
- text: il testo.
- org: le coordinate dell'angolo in basso a sinistra della stringa.
- font: il tipo di font; nel sotro caso cv2.FONT_HERSHEY_SIMPLEX
- fontScale: la misura del font.
- color: il colore il BRG.
- thickness: con esso indichiamo lo spessore della stringa.<br>

Inseriremo il testo della predizione nell'immagine sotto analisi, dunque tramite *cvutils.imshow()* mostreremo il risultato della previsione su jupyter.<br> (variante: *cv2.imshow()* che utilizza il terminale del pc per mostrare l'output) <br><br>Ecco le immagini date in input come test:</font><br> <img src="./test-ima.png" width="60%"/><br><br><br>

<br><br><font size="3">Ecco il codice appena discusso precedentemente:</font><br>

```python
test_path = "textH/dataset/test"
print("Results: ")
for file in glob.glob(test_path + "/*.jpg"):
    # read the input image
    image = cv2.imread(file)

    # convert to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # extract haralick texture from the image
    features = extract_features(gray)

    # evaluate the model and predict label
    prediction = clf_svm.predict(features.reshape(1, -1))[0]

    # show the label
    cv2.putText(image, prediction, (20,30), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0,255,255), 3)
    print("Prediction - {}".format(prediction))

    # display the output image
    # cv2.imshow(image) - Open local terminal
    cvutils.imshow("Prediction - {}".format(prediction), image)
    cv2.waitKey(0)
```

# 2. Local Binary Patterns<br>
<font size="3">Tramite le Local Binary Patterns è possibile effettuare una classificazione delle texture allenando un classificatore su un training set di texture note, usando come feature gli istogrammi di LBP estratti dalla texture.<img src="./lbp.png"><br><br>A differenza delle features di texture di Haralick che calcolano una rappresentazione globale di texture basata sulla matrice di ricorrenza del livello di grigio, gli LBP calcolano invece una  rappresentazione locale di texture. Questa rappresentazione locale viene costruita confrontando ciascun pixel con i dintorni circostanti di pixel.

Il primo passo per costruire il descrittore di texture LBP è convertire l'immagine in scala di grigi. Per ogni pixel nell'immagine in scala di grigi, selezioniamo un quartiere di dimensioni che circonda il pixel centrale. Un valore LBP viene quindi calcolato per questo pixel centrale e memorizzato nell'array 2D di output con la stessa larghezza e altezza dell'immagine di input.

Ad esempio, diamo un'occhiata al descrittore LBP originale che opera su un vicinato fisso, di pixel 3 x 3, in questo modo:</font><br><br><br><br><img src="lbp_mat.png" width="40%">


<font size="3">Se l'intensità del pixel centrale è maggiore o uguale al suo vicino, impostiamo il valore a 1; altrimenti, lo impostiamo a 0.<br><br>
Il  più grande svantaggio dell'algoritmo è l'impossibilità di acquisire dettagli su scale diverse, ma solamente su scala fissa  3 x 3!<br>

Per gestire ciò hanno proposto un'estensione dell'implementazione originale di LBP. Per tenere conto delle dimensioni variabili del vicinato, sono stati introdotti due parametri:

Il numero di punti p in un quartiere circolare simmetrico da considerare (eliminando così il ricorso a un quartiere quadrato).
Il raggio del cerchio r, che ci consente di tenere conto di scale diverse.</font><img src="exte_lbp.png" width="50%">

<font size="3">**OpenCV** implementa anche LBP, ma rigorosamente nel contesto del riconoscimento facciale.</font>

<br><br><font size="3">Inziamo creando un file *class_train.txt* avente una colonna contenente i path delle immagini di training e una colonna con le rispettive etichette (nel nostro esempio stiamo utilizzando 2 immagini per ogni classe):</font><img src="./class_train.png" width="20%">

<br><font size="3">Procediamo importanto le librerie d'interesse:</font>

```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
from skimage.feature import local_binary_pattern
from scipy.stats import itemfreq
from sklearn.preprocessing import normalize
import cvutils
import csv
```

<font size="3">Scriviamo il codice per memorizzare il percorso di tutte le immagini nel set di addestramento.</font>

```python
train_images = cvutils.imlist("lbp-texture/train/")
# Dictionary containing image paths as keys and corresponding label as value
train_dic = {}
with open('lbp-texture/class_train.txt', 'r') as csvfile:
    reader = csv.reader(csvfile, delimiter=' ')
    for row in reader:
        train_dic[row[0]] = int(row[1])
```

<font size="3">- In *train_images* memorizziamo tutti i percorsi delle immagini di allenamento.<br>- Creiamo un dizionario *train_dic* che conterrà il nome dell'immagine e l'etichetta della classe corrispondente.<br>- Successivamente leggiamo le righe del documento class_train.txt, sopra descritto, e quindi aggiungiamo la coppia chiave-valore a train_dic.<br> In questo tutorial, abbiamo suddiviso le nostre texture in due set: un training set di 2 immagini per classe (3 texture x 2 immagini  = 6 immagini totali) e un test set di una o due immagini per texture.<br><br>
Dopodichè passiamo al calcolo degli istogrammi LBP normalizzati, come segue:<br><br><br>Inizializziamo 3 liste:<br>
* X_test - per memorizzare gli istogrammi LBP normalizzati;
* X_name - per memorizzare l'indirizzo delle immagini;
* y_test - per memorizzare l'etichetta di classe corrispondente.
</font>
<img src="./test_ima_lbp.png" width="60%"><br><br>

```python
X_test = []
X_name = []
y_test = []
```

<font size="3">Adesso calcoliamo gli istogrammi per ogni immagine del training set, aggiornando le varie liste inizializzate poco prima:<br> 
- Quindi, in primo luogo leggiamo l'immagine corrente usando la funzione *cv2.imread()*. Quindi convertiamo l'immagine in scala di grigi poiché LBP funziona sull'immagine in scala di grigi. Successivamente, calcoliamo la maschera LBP, con *local_binary_pattern()*,dalla libreria *skimage*, che prende in input l'immagine in scala di grigi , il numero di punti (24) e il raggio (3). Come terzo parametro prende il tipo di motivo da utilizzare, in questo caso *uniform*, cioè il risultato include solo motivi in cui tutti i punti neri sono adiacenti e tutti i punti bianchi sono adiacenti.<br><img src="./uniform_lbp.png"><br>Una volta ottenuta la maschera, calcoliamo l'istogramma LBP e lo normalizziamo tramite *itemfreq()*, che prende come parametro l'array lbp , precedentemente calcolato, applicatogli la funzione *ravel()* che non fa altro che creare un array continuo.<br>Successivamente, aggiungiamo l'istogramma all'elenco X_test. Aggiungiamo il nome dell'immagine all'elenco X_name e l'etichetta della classe di immagini a y_test.<br><br>
- Infine visualizziamo le immagini di training per righe, tramite gli axes selezioniamo il riquadro di output per ogni immagine e tramite la funzione *imshow()* le mostriamo, inserendo un title per ognuna di essa:</font>

```python
for train_image in train_images:
    if train_image != "lbp-texture/train/.DS_Store":
        im = cv2.imread(train_image)
        # Convert to grayscale as LBP works on grayscale image
        im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
        radius = 3
        # Number of points to be considered as neighbourers 
        no_points = 8 * radius
        # Uniform LBP is used
        lbp = local_binary_pattern(im_gray, no_points, radius, method='uniform')
        # Calculate the histogram
        x = itemfreq(lbp.ravel())
        # Normalize the histogram
        hist = x[:, 1]/sum(x[:, 1])
        # Append image path in X_name
        X_name.append(train_image)
        # Append histogram to X_name
        X_test.append(hist)
        # Append class label in y_test
        y_test.append(train_dic[os.path.split(train_image)[1]])
    
# Display the training images
nrows = 2
ncols = 3
fig, axes = plt.subplots(nrows,ncols)
for row in range(nrows):
    for col in range(ncols):
        axes[row][col].imshow(cv2.cvtColor(cv2.imread(X_name[row*ncols+col]), cv2.COLOR_BGR2RGB))
        axes[row][col].axis('off')
        axes[row][col].set_title("{}".format(os.path.split(X_name[row*ncols+col])[1]))
```

<font size="3">Una volta che abbiamo estratto le nostre features ed etichette, possiamo alla fase di testing sul test set:
- Quindi calcoliamo gli istogrammi LBP normalizzati per ogni immagine nel set di test e quindi confrontiamo gli istogrammi LBP normalizzati delle immagini di allenamento (calcolati sopra) con questi usando la metrica della distanza Chi-Squared.<br><br>Ordiniamo quindi i risultati in base alla distanza Chi-Squared e visualizziamo i risultati in ordine.<br>Nel dettaglio: calcoliamo gli istogrammi LBP normalizzati come nel caso delle immagini di allenamento. Successivamente calcoliamo la distanza Chi-Squared dell'immagine di prova con tutte le immagini di allenamento utilizzando la funzione di OpenCV *cv2.compareHist()* che prende in input due istrogrammi e il metodo di confronto, (nel nostro caso cv2.HISTCMP_CHISQR).<br>Altri possibili metodi sono:<br><br>
- CV_COMP_CORREL Correlation;
- CV_COMP_CHISQR Chi-Square;
- CV_COMP_INTERSECT Intersection;
- CV_COMP_BHATTACHARYYA Bhattacharyya distance;
- CV_COMP_HELLINGER Synonym for CV_COMP_BHATTACHARYYA;<br><br>
Quindi, ordiniamo i punteggi, visualizziamo le immagini di training con il punteggio corrispondente: più bassa è la distanza Chi-Squared più l'immagine apparterrà a quella classe.</font><br>

```python
test_images = cvutils.imlist("lbp-texture/test/")
for test_image in test_images:
    print(test_image)
    if test_image != "lbp-texture/test/.DS_Store":
        im = cv2.imread(test_image)
        im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)
        radius = 3
        no_points = 8 * radius
        lbp = local_binary_pattern(im_gray, no_points, radius, method='uniform')
        x = itemfreq(lbp.ravel())
        hist = x[:, 1]/sum(x[:, 1])
        cvutils.imshow("** Query Image -> {}**".format(test_image), im)
        results = []

        for index, x in enumerate(X_test):
            score = cv2.compareHist(np.array(x, dtype=np.float32), np.array(hist, dtype=np.float32), cv2.HISTCMP_CHISQR)
            results.append((X_name[index], round(score, 3)))
        results = sorted(results, key=lambda score: score[1])
        # Display the results
        nrows = 2
        ncols = 3
        fig, axes = plt.subplots(nrows,ncols)
        fig.suptitle("** Scores for -> {}**".format(test_image))
        for row in range(nrows):
            for col in range(ncols):
                axes[row][col].imshow(cv2.cvtColor(cv2.imread(results[row*ncols+col][0]), cv2.COLOR_BGR2RGB))
                axes[row][col].axis('off')
                axes[row][col].set_title("Score {}".format(results[row*ncols+col][1]))
```

<font size="3">La fase di **sintesi** fa parte dei task principali delle texture: *sintetizzare* una texture significa che dato un campione di una texture, si può creare una nuova texture che se osservata da un essere umano, viene percepita come se fosse stata generata dallo stesso processo stocastico di quella del campione.<br> Questa tecnica può essere utile per svariati motivi, come per esempio la *compressione*, salvando il modello anzichè un'intera immagine e generarla all'occorrenza.<img src="./synthetis.png" width="40%"><br><br><br>Adesso procediamo con l'implementazione di un'algoritmo noto di *synthesis*: Efros e Freeman<br></font>

Inizializziamo un patch random per iniziare il processo di sintesi, procedendo con l'implementazione di due cicli for che scorrerrano per le dimensione del *PatchSize* scelto in precedenza, inziando a generare l'immagine di output:<br>

<br><font size="3">Adesso procediamo con l'implementazione di funzioni utili per l'avanzamento del processo di sintesi:<br>
- Tramite le prime due funzioni, calcoliamo l'errore di sovrapposizione verticale e orizzontale, che ci serviranno all'interno della funzione *GetBestPatches* per comprendere se il pixel in quesitone può far parte dei pixels ottimali, in tal caso lo mettiamo in una lista che poi sarà ritornata, alla fine dell'esecuzione di quest'ultima;<br>
- *GetBestPatches* prende in input un range di pixel e va a controllare i migliori patches da creare, partendo dal livello superiore (in alto), proseguendo con il livello più a sinistra, calcolando l'errore orizzontalmente e non più verticalmente e controllando anche i pixel centrali. Infine ritorna la lista dei migliori patches.
</font>

```python
def OverlapErrorVertical( imgPx, samplePx ):
    iLeft,jLeft = imgPx
    iRight,jRight = samplePx
    OverlapErr = 0
    diff = np.zeros((3))
    for i in range( PatchSize ):
        for j in range( OverlapWidth ):
            diff[0] =  int(img[i + iLeft, j+ jLeft][0]) - int(img_sample[i + iRight, j + jRight][0])
            diff[1] =  int(img[i + iLeft, j+ jLeft][1]) - int(img_sample[i + iRight, j + jRight][1])
            diff[2] =  int(img[i + iLeft, j+ jLeft][2]) - int(img_sample[i + iRight, j + jRight][2])
            OverlapErr += (diff[0]**2 + diff[1]**2 + diff[2]**2)**0.5
    return OverlapErr

def OverlapErrorHorizntl( leftPx, rightPx ):
    iLeft,jLeft = leftPx
    iRight,jRight = rightPx
    OverlapErr = 0
    diff = np.zeros((3))
    for i in range( OverlapWidth ):
        for j in range( PatchSize ):
            diff[0] =  int(img[i + iLeft, j+ jLeft][0]) - int(img_sample[i + iRight, j + jRight][0])
            diff[1] =  int(img[i + iLeft, j+ jLeft][1]) - int(img_sample[i + iRight, j + jRight][1])
            diff[2] =  int(img[i + iLeft, j+ jLeft][2]) - int(img_sample[i + iRight, j + jRight][2])
            OverlapErr += (diff[0]**2 + diff[1]**2 + diff[2]**2)**0.5
    return OverlapErr

def GetBestPatches( px ):
    PixelList = []
    #check for top layer
    if px[0] == 0:
        for i in range(sample_height - PatchSize):
            for j in range(OverlapWidth, sample_width - PatchSize ):
                error = OverlapErrorVertical( (px[0], px[1] - OverlapWidth), (i, j - OverlapWidth)  )
                if error  < ThresholdOverlapError:
                    PixelList.append((i,j))
                elif error < ThresholdOverlapError/2:
                    return [(i,j)]
    #check for leftmost layer
    elif px[1] == 0:
        for i in range(OverlapWidth, sample_height - PatchSize ):
            for j in range(sample_width - PatchSize):
                error = OverlapErrorHorizntl( (px[0] - OverlapWidth, px[1]), (i - OverlapWidth, j)  )
                if error  < ThresholdOverlapError:
                    PixelList.append((i,j))
                elif error < ThresholdOverlapError/2:
                    return [(i,j)]
    #for pixel placed inside 
    else:
        for i in range(OverlapWidth, sample_height - PatchSize):
            for j in range(OverlapWidth, sample_width - PatchSize):
                error_Vertical   = OverlapErrorVertical( (px[0], px[1] - OverlapWidth), (i,j - OverlapWidth)  )
                error_Horizntl   = OverlapErrorHorizntl( (px[0] - OverlapWidth, px[1]), (i - OverlapWidth,j) )
                if error_Vertical  < ThresholdOverlapError and error_Horizntl < ThresholdOverlapError:
                    PixelList.append((i,j))
                elif error_Vertical < ThresholdOverlapError/2 and error_Horizntl < ThresholdOverlapError/2:
                    return [(i,j)]
    return PixelList
```

<font size="3">Successivamente con le seguenti implementazioni delle funzioni sottostanti andiamo a calcolare i costi in termini di larghezza e altezza tramite l'utilizzo della funzione *SSD_Error*.<br>Mentre con le funzioni *FindMinCostPathVertical* e *FindMinCostPathHorizntl* andiamo a cercare i patch ottimali, con il minor spreco di offset su cui andare ad effettuare l'operazione di *quilting*:</font>

```python
def SSD_Error( offset, imgPx, samplePx ):
    err_r = int(img[imgPx[0] + offset[0], imgPx[1] + offset[1]][0]) -int(img_sample[samplePx[0] + offset[0], samplePx[1] + offset[1]][0])
    err_g = int(img[imgPx[0] + offset[0], imgPx[1] + offset[1]][1]) - int(img_sample[samplePx[0] + offset[0], samplePx[1] + offset[1]][1])
    err_b = int(img[imgPx[0] + offset[0], imgPx[1] + offset[1]][2]) - int(img_sample[samplePx[0] + offset[0], samplePx[1] + offset[1]][2])
    return (err_r**2 + err_g**2 + err_b**2)/3.0

#---------------------------------------------------------------#
#|                  Calculating Cost                           |#
#---------------------------------------------------------------#

def GetCostVertical(imgPx, samplePx):
    Cost = np.zeros((PatchSize, OverlapWidth))
    for j in range(OverlapWidth):
        for i in range(PatchSize):
            if i == PatchSize - 1:
                Cost[i,j] = SSD_Error((i ,j - OverlapWidth), imgPx, samplePx)
            else:
                if j == 0 :
                    Cost[i,j] = SSD_Error((i , j - OverlapWidth), imgPx, samplePx) + min( SSD_Error((i + 1, j - OverlapWidth), imgPx, samplePx),SSD_Error((i + 1,j + 1 - OverlapWidth), imgPx, samplePx) )
                elif j == OverlapWidth - 1:
                    Cost[i,j] = SSD_Error((i, j - OverlapWidth), imgPx, samplePx) + min( SSD_Error((i + 1, j - OverlapWidth), imgPx, samplePx), SSD_Error((i + 1, j - 1 - OverlapWidth), imgPx, samplePx) )
                else:
                    Cost[i,j] = SSD_Error((i, j -OverlapWidth), imgPx, samplePx) + min(SSD_Error((i + 1, j - OverlapWidth), imgPx, samplePx),SSD_Error((i + 1, j + 1 - OverlapWidth), imgPx, samplePx),SSD_Error((i + 1, j - 1 - OverlapWidth), imgPx, samplePx))
    return Cost

def GetCostHorizntl(imgPx, samplePx):
    Cost = np.zeros((OverlapWidth, PatchSize))
    for i in range( OverlapWidth ):
        for j in range( PatchSize ):
            if j == PatchSize - 1:
                Cost[i,j] = SSD_Error((i - OverlapWidth, j), imgPx, samplePx)
            elif i == 0:
                Cost[i,j] = SSD_Error((i - OverlapWidth, j), imgPx, samplePx) + min(SSD_Error((i - OverlapWidth, j + 1), imgPx, samplePx), SSD_Error((i + 1 - OverlapWidth, j + 1), imgPx, samplePx))
            elif i == OverlapWidth - 1:
                Cost[i,j] = SSD_Error((i - OverlapWidth, j), imgPx, samplePx) + min(SSD_Error((i - OverlapWidth, j + 1), imgPx, samplePx), SSD_Error((i - 1 - OverlapWidth, j + 1), imgPx, samplePx))
            else:
                Cost[i,j] = SSD_Error((i - OverlapWidth, j), imgPx, samplePx) + min(SSD_Error((i - OverlapWidth, j + 1), imgPx, samplePx), SSD_Error((i + 1 - OverlapWidth, j + 1), imgPx, samplePx), SSD_Error((i - 1 - OverlapWidth, j + 1), imgPx, samplePx))
    return Cost

#---------------------------------------------------------------#
#|                  Finding Minimum Cost Path                  |#
#---------------------------------------------------------------#

def FindMinCostPathVertical(Cost):
    Boundary = np.zeros((PatchSize),np.int)
    ParentMatrix = np.zeros((PatchSize, OverlapWidth),np.int)
    for i in range(1, PatchSize):
        for j in range(OverlapWidth):
            if j == 0:
                ParentMatrix[i,j] = j if Cost[i-1,j] < Cost[i-1,j+1] else j+1
            elif j == OverlapWidth - 1:
                ParentMatrix[i,j] = j if Cost[i-1,j] < Cost[i-1,j-1] else j-1
            else:
                curr_min = j if Cost[i-1,j] < Cost[i-1,j-1] else j-1
                ParentMatrix[i,j] = curr_min if Cost[i-1,curr_min] < Cost[i-1,j+1] else j+1
            Cost[i,j] += Cost[i-1, ParentMatrix[i,j]]
    minIndex = 0
    for j in range(1,OverlapWidth):
        minIndex = minIndex if Cost[PatchSize - 1, minIndex] < Cost[PatchSize - 1, j] else j
    Boundary[PatchSize-1] = minIndex
    for i in range(PatchSize - 1,0,-1):
        Boundary[i - 1] = ParentMatrix[i,Boundary[i]]
    return Boundary

def FindMinCostPathHorizntl(Cost):
    Boundary = np.zeros(( PatchSize),np.int)
    ParentMatrix = np.zeros((OverlapWidth, PatchSize),np.int)
    for j in range(1, PatchSize):
        for i in range(OverlapWidth):
            if i == 0:
                ParentMatrix[i,j] = i if Cost[i,j-1] < Cost[i+1,j-1] else i + 1
            elif i == OverlapWidth - 1:
                ParentMatrix[i,j] = i if Cost[i,j-1] < Cost[i-1,j-1] else i - 1
            else:
                curr_min = i if Cost[i,j-1] < Cost[i-1,j-1] else i - 1
                ParentMatrix[i,j] = curr_min if Cost[curr_min,j-1] < Cost[i-1,j-1] else i + 1
            Cost[i,j] += Cost[ParentMatrix[i,j], j-1]
    minIndex = 0
    for i in range(1,OverlapWidth):
        minIndex = minIndex if Cost[minIndex, PatchSize - 1] < Cost[i, PatchSize - 1] else i
    Boundary[PatchSize-1] = minIndex
    for j in range(PatchSize - 1,0,-1):
        Boundary[j - 1] = ParentMatrix[Boundary[j],j]
    return Boundary

```

<font size="3">Qui troviamo le funzioni per l'operazione di *Quilting*:<br> 
- *QuiltVertical* e *QuiltHorizntl* svolgono la vera e propria fase di quilting in verticale ed orizzontale;<br>
- la terza funzione *QuiltPatches* inizia la fase di quilting dal punto in alto a sinistra ( il primo if) prosegue con il quilting orizzontale (elif) ed infine per i pixel interni (else); tramite le funzioni *GetCostVertical* e *GetCostHorizntl* andiamo a ricavare il costo passandolo poi come parametro alle funzioni *FindMinCostPathVertical* e *FindMinCostPathHorizontal* le quali andranno a ricavare i bordi da "ricamare", ovvero i bordi da dove iniziare la fase di quilting.</font>

```python
#---------------------------------------------------------------#
#|                      Quilting                               |#
#---------------------------------------------------------------#

def QuiltVertical(Boundary, imgPx, samplePx):
    for i in range(PatchSize):
        for j in range(Boundary[i], 0, -1):
            img[imgPx[0] + i, imgPx[1] - j] = img_sample[ samplePx[0] + i, samplePx[1] - j ]
def QuiltHorizntl(Boundary, imgPx, samplePx):
    for j in range(PatchSize):
        for i in range(Boundary[j], 0, -1):
            img[imgPx[0] - i, imgPx[1] + j] = img_sample[samplePx[0] - i, samplePx[1] + j]

def QuiltPatches( imgPx, samplePx ):

    if imgPx[0] == 0:
        Cost = GetCostVertical(imgPx, samplePx)
        Boundary = FindMinCostPathVertical(Cost)
        QuiltVertical(Boundary, imgPx, samplePx)
    elif imgPx[1] == 0:
        Cost = GetCostHorizntl(imgPx, samplePx)
        Boundary = FindMinCostPathHorizntl(Cost)
        QuiltHorizntl(Boundary, imgPx, samplePx)
    else:
        CostVertical = GetCostVertical(imgPx, samplePx)
        CostHorizntl = GetCostHorizntl(imgPx, samplePx)
        BoundaryVertical = FindMinCostPathVertical(CostVertical)
        BoundaryHorizntl = FindMinCostPathHorizntl(CostHorizntl)
        QuiltVertical(BoundaryVertical, imgPx, samplePx)
        QuiltHorizntl(BoundaryHorizntl, imgPx, samplePx)

#Growing Image Patch-by-patch
def FillImage( imgPx, samplePx ):
    for i in range(PatchSize):
        for j in range(PatchSize):
            img[ imgPx[0] + i, imgPx[1] + j ] = img_sample[ samplePx[0] + i, samplePx[1] + j ]

```

<font size="3">Infine, tramite la libreria *sys* andiamo a tenere traccia del progresso in corso; successivamente troviamo due cicli while il primo dei quali terrà traccia della fine dell'immagine di output in altezza; inizializziamo quindi il progresso a 0 affinchè ciclerà (il secondo ciclo) fino a quando non verranno trovate le migliori corrispondenze per il pixel corrente. Il secondo ciclo, invece, andrà a calcolare le migliori corrispondenze per il pixel corrente tramite la funzione che abbiamo visto in precedenza *GetBestPatches*. Se quest'ultime sono presenti, allora, entriamo nell'if, in cui viene calcolato una selezione casuale dai pixels più adatti, utilizzando la funzione *randint* all'interno della lista delle migliori corrispondenze, dopodichè si passa alla fase di quilting.<br>Se non vi sono corrispondenze ottimali si procede aumentando la threshold, impostata all'inizio del codice, come vediamo nell'istruzione else.<br> Infine, si mostra l'immagine finale insieme alla texture presa in input tramite le funzione *imshow* di *cvutils* spiegata dettagliatamente in precedenza.  </font>

```python
pixelsCompleted = 0
TotalPatches = ( (img_height - 1 )/ PatchSize )*((img_width)/ PatchSize) - 1
sys.stdout.write("Progress : [%-20s] %d%% | PixelsCompleted: %d | ThresholdConstant: --.------" % ((pixelsCompleted*20/TotalPatches), (100*pixelsCompleted)/TotalPatches, pixelsCompleted))
sys.stdout.flush()
while GrowPatchLocation[0] + PatchSize < img_height:
    pixelsCompleted += 1
    ThresholdConstant = InitialThresConstant
    progress = 0 
    while progress == 0:
        ThresholdOverlapError = ThresholdConstant * PatchSize * OverlapWidth
        #Get Best matches for current pixel
        List = GetBestPatches(GrowPatchLocation)
        if len(List) > 0:
            progress = 1
            sampleMatch = List[ randint(0, len(List) - 1) ]
            FillImage( GrowPatchLocation, sampleMatch )
            #Quilt this with in curr location
            QuiltPatches( GrowPatchLocation, sampleMatch )
            #Update cur pixel location
            GrowPatchLocation = (GrowPatchLocation[0], GrowPatchLocation[1] + PatchSize)
            if GrowPatchLocation[1] + PatchSize > img_width:
                GrowPatchLocation = (GrowPatchLocation[0] + PatchSize, 0)
        #if not progressed, increse threshold
        else:
            ThresholdConstant *= 1.1
    sys.stdout.write('\r')
    sys.stdout.write("Progress : [%-20s] %d%% | PixelsCompleted: %d | ThresholdConstant: %f" % ((pixelsCompleted*20/TotalPatches), (100*pixelsCompleted)/TotalPatches, pixelsCompleted, ThresholdConstant))
    sys.stdout.flush()
    
cvutils.imshow('Sample Texture',img_sample)
cv2.waitKey(0)
cv2.destroyAllWindows()
cvutils.imshow('Generated Image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
